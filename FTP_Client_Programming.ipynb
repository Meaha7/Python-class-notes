# cs.brown.edu's ftp site

```python
from ftplib import FTP
```

```python
f = FTP('ftp.cs.brown.edu')
```

```python
f.login()
```

```python
f.dir()
```

```python
f.pwd()
```

```python
f.cwd('/incoming')
```

```python
f.dir()
```

```python
f.storlines('STOR SS1.txt', open('/content/sample.txt', 'rb'))
#Create SampleTextFile.txt in current working directory
```

```python
f.dir()
```

```python
f.retrlines('RETR SS1.txt', open('Sample.txt', 'w').write)
```

```python
f.quit()
```
